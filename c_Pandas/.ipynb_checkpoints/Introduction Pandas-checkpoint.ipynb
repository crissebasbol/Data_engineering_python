# Series 

```python
import pandas as pd

series_test = pd.Series([100, 200, 300])

series_test
```

```python
series_test2 = pd.Series({1999: 48,
                          2000: 65,
                          2001: 89})

series_test2
```

# DataFrames

```python
frame_test = pd.DataFrame({1999: [47, 38, 39],
                           2000: [34, 32, 32],
                           2001: [23, 39, 23]})
                           
frame_test
```

```python
frame_test2 = pd.DataFrame([[47, 38, 39],
                          [34, 32, 32],
                          [23, 39, 23]], columns=[1999, 2000, 2001])

frame_test2
```
